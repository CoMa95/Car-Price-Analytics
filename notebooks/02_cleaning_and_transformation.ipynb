## Remove Duplicates

## Handle Typos

First we will create manufacturer and model from CarName and Drop CarName.

```python
df["manufacturer"] = df["CarName"].apply(lambda x: x.split(" ")[0]) # Extract manufacturer from CarName
df["model"] = df["CarName"].apply(lambda x: " ".join(x.split(" ")[1:])) # Extract model from CarName

df["model"].fillna(df["manufacturer"], inplace=True) # Fill missing model values with manufacturer name

df.drop("CarName", axis=1, inplace=True) # Drop the original CarName column
print("‚úÖ Encoded 'manufacturer' from 'CarName' and dropped 'CarName' column.") # Confirmation message
```

```python
df['manufacturer'] = df['manufacturer'].str.lower() # Convert manufacturer names to lowercase for consistency
for manufacturer in df["manufacturer"].unique(): # Iterate over unique manufacturers
    print(f" - {manufacturer}") # Display unique manufacturers
```

we can see from the above ist we can see a number of typos for the manufacturer (although 'Nissan' is technically not a typo it will be c). These will no be corrected.

```python
df['manufacturer'] = df['manufacturer'].replace({'maxda': 'mazda',
                                       'porcshce': 'porsche',
                                       'toyouta': 'toyota',
                                       'vokswagen': 'volkswagen',
                                       'vw': 'volkswagen', 
                                       'Nissan': 'nissan'}) # Correct typos in manufacturer names
print("‚úÖ Corrected typos in 'manufacturer' names.") # Confirmation message
for manufacturer in df["manufacturer"].unique(): # Iterate over unique manufacturers
    print(f" - {manufacturer}") # Display unique manufacturers
```

We now have the correct list of manufacturers.

## Encode Categorical Variables

change fueltype gas to petrol

```python
df['fueltype'] = df['fueltype'].replace('gas', 'petrol')
print("‚úÖ Standardised 'fueltype' values from 'gas' to 'petrol'.")
```

combine `carbody` values `hardtop` and `convertible`

```python
df['carbody'] = df['carbody'].replace({'hardtop': 'convertible'})
print("‚úÖ Combined 'carbody' values 'hardtop' and 'convertible'.")
```

```python
categorical_columns = df.select_dtypes(include=['object']).columns.tolist() # Identify categorical columns  
df[categorical_columns] = df[categorical_columns].astype('category') # Convert categorical columns to 'category' dtype

print("‚úÖ All categorical variables set to 'category' dtype for readability.")  # Confirmation message
display(df[categorical_columns].dtypes) # Display the data types of categorical columns to confirm conversion
display(df[categorical_columns].head()) # Display the first few rows of categorical columns
```

---

## Drop uneeded columns

```python
columns_to_drop = ['car_ID', 'enginelocation'] # Define columns to drop

df.drop(columns=columns_to_drop, inplace=True) # Drop the specified columns
print("‚úÖ Dropped unnecessary columns for analysis.") # Confirmation message
```
